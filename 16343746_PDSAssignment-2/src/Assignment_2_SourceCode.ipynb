```python
import pandas as pd
```

```python
# Reading the csv data file
file_path = "/content/sample_data/train.csv"
data = pd.read_csv(file_path)
```

```python
data
```

```python
#TASK A
# Looking for the missing values in all the columns
missing_vals = data.isnull().sum()
missing_vals
```

```python
missing_vars = list(missing_vals[missing_vals > 0].index)
missing_vars
```

```python
# Removing columns with more than 50% of missing values
data = data.dropna(thresh=0.5, axis=1)
```

```python
# Imputing the missing values in remaining columns with the median
impute_vals = data.median(axis=0)
data = data.fillna(impute_vals)
```

```python
# Storing the updated Data into Clean data csv file
data.to_csv('/content/clean_data/clean_data.csv', index=False)
data
```

```python
#TASK B
# Reading the clean csv data file
file_path = '/content/clean_data/clean_data.csv'
clean_data = pd.read_csv(file_path)

# Specifing the columns to clean
cols_to_clean = ['Mileage', 'Engine', 'Power', 'New_Price']

# Cleaning each column
for col in cols_to_clean:
    # Removing non-numerical characters
    clean_data[col] = clean_data[col].str.replace('[^\d.]', '')

# Changing the cleaned columns to numeric
clean_data[cols_to_clean] = clean_data[cols_to_clean].apply(pd.to_numeric)

# Storing the updated Data into the numerical_data csv file
clean_data.to_csv('/content/results/numerical_data.csv', index=False)
clean_data
```

```python
#TASK C
#Performing one hot encoding
one_hot_encoded_value = pd.get_dummies(clean_data, columns=['Fuel_Type', 'Transmission'])
one_hot_encoded_value.to_csv('/content/results/one_hot_encoded_value.csv', index=False)
one_hot_encoded_value
```

```python
#TASK D
# Printing the current year
current_year = pd.to_datetime('today').year

# Calculating the age of the car
one_hot_encoded_value['Age'] = current_year - one_hot_encoded_value['Year']

# Storing the updated Data into the car_age_data csv file
one_hot_encoded_value.to_csv('/content/results/car_age_data.csv', index=False)
one_hot_encoded_value
```

```python
#TASK E
# Select operation
select_data = one_hot_encoded_value[['Year' , 'Kilometers_Driven' , 'Owner_Type' , 'New_Price']]
print("Select data:\n", select_data)
```

```python
# Filtering rows
filter_data = one_hot_encoded_value[one_hot_encoded_value['Year'] > 2009]
print("Filtered data:\n", filter_data)
```

```python
# Renaming columns
rename_data = one_hot_encoded_value.rename(columns={'New_Price': 'Current_Price', 'Power': 'HorsePower'})
print("Renamed data:\n", rename_data)
```

```python
# Mutating the data
mutate_data = one_hot_encoded_value.assign(Performance=one_hot_encoded_value['Kilometers_Driven'] / one_hot_encoded_value['Mileage'])
print("Mutated data:\n", mutate_data)
```

```python
# Arranging the data
sorted_data = one_hot_encoded_value.sort_values(by='Kilometers_Driven', ascending=True)
print("Arranged data:\n", sorted_data)
```

```python
# Summarizing the data with group by function
grouped_data = one_hot_encoded_value.groupby('Owner_Type').agg({'Mileage': 'median', 'Kilometers_Driven': 'mean'})
print("Summarized data:\n", grouped_data)
```
